# LunarLander

```python
env = gym.make("LunarLander-v2", render_mode="human")
observation, info = env.reset(seed=42)
for _ in range(1000):
    action = env.action_space.sample()  # this is where you would insert your policy
    observation, reward, terminated, truncated, info = env.step(action)
    print(terminated, truncated, reward, info)
    if terminated or truncated:
        observation, info = env.reset()

env.close()
```

# CartPole

```python
env = gym.make('CartPole-v1', render_mode="human")
```

```python
done = False
score = 0
state = env.reset()

while not done:
    env.render()
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)
    if terminated or truncated:
        done=True
    score += reward

print('total reward:', score)
```

```python
for epoch in range(1, 6):
    done = False
    score = 0
    state = env.reset()
    while not done:
        env.render()
        action = env.action_space.sample()
        observation, reward, terminated, truncated, info = env.step(action)
        if terminated or truncated:
            done=True
        score += reward
    print(f'epoch: {epoch}, total reward: {score}')
```

## render save

```python
import gymnasium as gym
import time

env = gym.make('CartPole-v1', render_mode="rgb_array")

done = False
score = 0
state = env.reset()

frames = []
while not done:
    frames.append(env.render())
    # env.render()
    action = env.action_space.sample()
    print(action)
    print(frames)
    observation, reward, terminated, truncated, info = env.step(action)
    if terminated or truncated:
        done=True
    score += reward
    time.sleep(0.2)

env.close()
print('total reward:', score)
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation

def display_frames_to_video(frames):
    plt.figure(figsize=(frames[0].shape[0] / 72.0, frames[0].shape[1] / 72.0), dpi=72)
    patch = plt.imshow(frames[0])

    def animate(i):
        patch.set_data(frames[i])

    ani = animation.FuncAnimation(plt.gcf(), animate, frames=range(len(frames)), interval=50)
    ani.save('cartpole.gif', writer='ffmpeg') # , writer='ffmpeg', fps=60

display_frames_to_video(frames)
```
